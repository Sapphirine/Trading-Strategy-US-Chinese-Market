```python
import pandas as pd
import numpy as np
import datetime
from universal import tools
from universal import algos
def transfer(date):
    d,m,y = date.split('/')
    y,m,d = int(y), int(m), int(d)
    return datetime.datetime(y,m,d)
```

```python
def get_benchmark(data,label):
    OLMAR = algos.OLMAR(window=5, eps=10)
    OLMAR.PRICE_TYPE = 'raw'
    result = OLMAR.run(data)
    print(result.summary())
    OLMAR_result = (result._X* result._B).sum(axis = 1).cumprod()

    PAMR = algos.PAMR()
    PAMR.PRICE_TYPE = 'raw'
    result = PAMR.run(data)
    print(result.summary())
    PAMR_result = (result._X* result._B).sum(axis = 1).cumprod()

    WMAMR = algos.WMAMR(window=5, eps=10)
    WMAMR.PRICE_TYPE = 'raw'
    result = WMAMR.run(data)
    print(result.summary())
    WMAMR_result = (result._X* result._B).sum(axis = 1).cumprod()

    BAH = algos.BAH()
    BAH.PRICE_TYPE = 'raw'
    result = BAH.run(data)
    print(result.summary())
    BAH_result = (result._X* result._B).sum(axis = 1).cumprod()
    benchmark = pd.DataFrame({"OLMAR":OLMAR_result,"PAMR":PAMR_result,"WMAMR":WMAMR_result,"BAH":BAH_result})
    benchmark.to_csv("benchmark_{}.csv".format(label))
    return OLMAR_result, PAMR_result, WMAMR_result,BAH_result
```

```python

HS300 = pd.read_csv("result_hs300.csv")
data = HS300[HS300.columns.difference(['test_date'])]
OLMAR_result, PAMR_result, WMAMR_result,BAH_result = get_benchmark(data,'HS300')
```

```python
def backtest(Data,mode='price'):
    if mode == 'signal':
        position = {'Buy':0,'Sell':0,'Hold':0}
        place= []
        money = 0
        hold = []
        for i in range(Data.shape[0]-1):
            if i > 3 and len(hold)>3 and hold[-1] < hold[-2] and hold[-2] <  2.0* np.std(hold[0:-3]):
                place += ['Hold']
                if position['Sell'] > 0:
                    position['Sell'] = 0
                    money += -Data['true_value'][i]
                elif position['Buy'] > 0:
                    position['Buy'] = 0
                    money += Data['true_value'][i]
                hold += [money + position['Buy'] * Data['true_value'][i+1] - position['Sell'] * Data['true_value'][i+1]]
            
                continue
            if Data['dqn_action'][i] == 'Buy':
                place += ['Buy']
                if position['Buy'] > 0:
                    money += 0
                else:
                    if position['Sell'] > 0:
                        position['Sell'] = 0
                        position['Hold'] = 0
                        money += -Data['true_value'][i]
                    else:
                        position['Buy'] = 1
                        money += -Data['true_value'][i]
            elif Data['dqn_action'][i] == 'Sell':
                place += ['Sell']
                if position['Sell'] > 0:
                    money += 0
                else:
                    if position['Buy'] > 0:
                        position['Buy'] = 0
                        position['Hold'] = 0
                        
                        money += 1*Data['true_value'][i]
                    else:
                        position['Sell'] = 1
                        money += Data['true_value'][i]
            
            else:
                place += ['Hold']
                if position['Sell'] > 0:
                    position['Sell'] = 0
                    money += -Data['true_value'][i]
                elif position['Buy'] > 0:
                    position['Buy'] = 0
                    money += Data['true_value'][i] 
            hold += [money + position['Buy'] * Data['true_value'][i+1] - position['Sell'] * Data['true_value'][i+1]]
        return hold,place
    if mode == 'price':   
        position = {'Buy':0,'Sell':0,'Hold':0}
        money = 0
        hold = []
        place= []
        for i in range(1,Data.shape[0]):

            if i > 3 and len(hold)>3 and hold[-1] < hold[-2] and hold[-2] <  2.0* np.std(hold[0:-3]):
                place += ['Hold']
                if position['Sell'] > 0:
                    position['Sell'] = 0
                    money += -Data['true_value'][i-1]
                elif position['Buy'] > 0:
                    position['Buy'] = 0
                    money += Data['true_value'][i-1]
                hold += [money + position['Buy'] * Data['true_value'][i] - position['Sell'] * Data['true_value'][i]]
                continue

            if Data['prediction'][i] > Data['prediction'][i-1]:
                place += ['Buy']
                if position['Buy'] > 0:
                      money += 0
                else:
                    if position['Sell'] > 0:
                        position['Sell'] = 0
                        position['Buy'] = 1
                        money += -2*Data['true_value'][i-1]
                    else:
                        position['Buy'] = 1
                        money += -Data['true_value'][i-1]

            elif Data['prediction'][i] < Data['prediction'][i-1] :
                place += ['Sell']
                if position['Sell'] > 0:
                      money += 0
                else:
                    if position['Buy'] > 0:
                        position['Buy'] = 0
                        position['Sell'] = 1
                        money += 2*Data['true_value'][i-1] 
                    else:
                        position['Sell'] = 1
                        money += Data['true_value'][i-1]
            else:
                place += ['Hold']
                if position['Sell'] > 0:
                    position['Sell'] = 0
                    money += -Data['true_value'][i-1]
                elif position['Buy'] > 0:
                    position['Buy'] = 0
                    money += Data['true_value'][i-1]
            hold += [money + position['Buy'] * Data['true_value'][i] - position['Sell'] * Data['true_value'][i]]
        #pd.DataFrame(hold).plot()
        return hold,place
    else:
        print("Woops!, The parameter mode must be in signal or price")
        
```
